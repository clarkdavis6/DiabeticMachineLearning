```python
import pandas as pd
import matplotlib
from matplotlib import pyplot
```

```python
file = "../data/dexcomData.csv"
```

```python
def parser(x):
    return pd.datetime.strptime('190'+x, '%Y-%m')
```

```python
glucose = pd.read_csv(file, parse_dates=[0])
```

```python
glucose['Blood Glucose (mg/dL)'] = glucose['Blood Glucose (mg/dL)'].astype('float')
```

```python
glucose = glucose[['Timestamp', 'Blood Glucose (mg/dL)']]
```

```python
glucose['Timestamp'] = pd.to_datetime(glucose['Timestamp'])
```

```python
glucose.set_index('Timestamp')
```

```python
gplot = glucose.plot(kind = 'line',
                     x = 'Timestamp',
                     y = 'Blood Glucose (mg/dL)',
                     figsize = (22, 12),
                     use_index = 1,
                     title = 'Glucose Readings')
gplot.set_xlabel('Month / Year')
gplot.set_ylabel('Blood Glucose (mg/dL)')
```

```python
pyplot.show()
```
